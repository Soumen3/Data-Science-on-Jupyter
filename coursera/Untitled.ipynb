```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.DataFrame({
    'Year': [2000, 2001, 2002, 2003, 2004, 2000, 2001, 2002, 2003, 2004],
    'Sales': [500, 700, 800, 650, 900, 600, 800, 850, 700, 950],
    'Vehicle Type': ['Car', 'Car', 'Car', 'Car', 'Car', 'Truck', 'Truck', 'Truck', 'Truck', 'Truck']
})

# Define recession years and non-recession years
recession_years = list(range(2007, 2009))  # adjust this according to your data
df['Period'] = df['Year'].apply(lambda x: 'Recession' if x in recession_years else 'Non-Recession')

# Group by 'Period', 'Year' and 'Vehicle Type' and sum 'Sales'
sales_per_year = df.groupby(['Period', 'Year', 'Vehicle Type'])['Sales'].sum().reset_index()

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(data=sales_per_year, x='Year', y='Sales', hue='Vehicle Type')
plt.title('Automobile Sales Over the Years by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'df' with columns 'Ad_Expenditure' and 'Period'
df = pd.DataFrame({
	'Ad_Expenditure': [50000, 70000],
	'Period': ['Recession', 'Non-Recession']
})

# Calculate the total advertising expenditure for each period
ad_exp = df.groupby('Period')['Ad_Expenditure'].sum()

plt.pie(ad_exp, labels=ad_exp.index, autopct='%1.1f%%')

plt.title('Portion of Advertising Expenditure During Recession and Non-Recession Periods')
plt.show()
```

```python
import dash
from dash.dependencies import Input, Output
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
	'Year': [2005, 2006, 2007, 2008, 2009, 2005, 2006, 2007, 2008, 2009],
	'Unemployment_Rate': [5.1, 4.6, 4.6, 5.8, 9.3, 5.1, 4.6, 4.6, 5.8, 9.3],
	'Sales': [500, 700, 800, 650, 900, 600, 800, 850, 700, 950],
	'Vehicle_Type': ['Car', 'Car', 'Car', 'Car', 'Car', 'Truck', 'Truck', 'Truck', 'Truck', 'Truck'],
	'Period': ['Recession', 'Recession', 'Recession', 'Recession', 'Recession', 'Recession', 'Recession', 'Recession', 'Recession', 'Recession']
})

# Create a Dash application
app = dash.Dash(__name__)

# Define the layout of the application
# Define the layout of the application
# Define the layout of the application
app.layout = html.Div(children=[
    html.H1(children='Recession Report Statistics Dashboard', style={'color': 'white'}),

    html.Div(children='''Dash: A web application framework for Python.''', style={'color': 'white'}),

    html.Label('Select Year'),
    dcc.Dropdown(
        id='year-dropdown',
        options=[{'label': i, 'value': i} for i in df['Year'].unique()],
        value=df['Year'].min()
    ),

    html.Label('Select Vehicle Type'),
    dcc.Dropdown(
        id='vehicle-type-dropdown',
        options=[{'label': i, 'value': i} for i in df['Vehicle_Type'].unique()],
        value='Car'
    ),

    html.Div(children=[
        dcc.Graph(id='sales-graph', style={'display': 'inline-block', 'width': '33%'}),
        dcc.Graph(id='unemployment-graph', style={'display': 'inline-block', 'width': '33%'}),
        dcc.Graph(id='sales-pie-chart', style={'display': 'inline-block', 'width': '33%'})
    ], style={'display': 'flex'})
])

# Define the callback for the sales graph (Bar Graph)
@app.callback(
    Output('sales-graph', 'figure'),
    [Input('year-dropdown', 'value'),
     Input('vehicle-type-dropdown', 'value')]
)
def update_sales_graph(selected_year, selected_vehicle_type):
    filtered_df = df[(df.Year == selected_year) & (df.Vehicle_Type == selected_vehicle_type)]
    figure = {
        'data': [
            {'x': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'bar', 'name': 'Car'},
            {'x': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'bar', 'name': 'Truck'},
        ],
        'layout': {
            'title': 'Automobile Sales During Recession Period'
        }
    }
    return figure

# Define the callback for the unemployment graph (Line Chart)
@app.callback(
    Output('unemployment-graph', 'figure'),
    [Input('year-dropdown', 'value'),
     Input('vehicle-type-dropdown', 'value')]
)
def update_unemployment_graph(selected_year, selected_vehicle_type):
    filtered_df = df[(df.Year == selected_year) & (df.Vehicle_Type == selected_vehicle_type)]
    figure = {
        'data': [
            {'x': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'line', 'name': 'Car'},
            {'x': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'line', 'name': 'Truck'},
        ],
        'layout': {
            'title': 'Automobile Sales During Recession Period'
        }
    }
    return figure

# Define the callback for the sales pie chart (Pie Chart)
@app.callback(
    Output('sales-pie-chart', 'figure'),
    [Input('year-dropdown', 'value'),
     Input('vehicle-type-dropdown', 'value')]
)
def update_sales_pie_chart(selected_year, selected_vehicle_type):
    filtered_df = df[(df.Year == selected_year) & (df.Vehicle_Type == selected_vehicle_type)]
    figure = {
        'data': [
            {'x': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Car') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'pie', 'name': 'Car'},
            {'x': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Year'], 
             'y': df[(df['Vehicle_Type'] == 'Truck') & (df['Year'] == selected_year)]['Sales'], 
             'type': 'pie', 'name': 'Truck'},
        ],
        'layout': {
            'title': 'Automobile Sales During Recession Period'
        }
    }
    return figure

if __name__ == '__main__':
    app.run_server(debug=True)
```
